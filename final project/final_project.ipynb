```python
'''
Пять моделей ML
'''
class TeamFormFeatures(BaseEstimator, TransformerMixin):
    """Трансформер для создания признаков формы команд с надежной обработкой NaN"""
    def __init__(self, window_sizes=[3, 5, 10]):
        self.window_sizes = window_sizes
        
    def fit(self, X, y=None):
        return self
        
    def transform(self, X):
        try:    
            X = X.copy()
            original_columns = set(X.columns)
            
            for team_col in ['HomeTeam', 'AwayTeam']:
                if team_col in X.columns:
                    for window in self.window_sizes:
                        goals = []
                        results = []
                        conceded = []
                        
                        # Создаем сдвинутые признаки
                        for i in range(1, window + 1):
                            # Для голов
                            goal_col = f'{team_col}_shift{i}_goals'
                            X[goal_col] = X.groupby(team_col)['FTHG' if team_col == 'HomeTeam' else 'FTAG'].shift(i)
                            goals.append(goal_col)
                            
                            # Для результатов (победа=1, ничья=0.5, поражение=0)
                            result_col = f'{team_col}_shift{i}_result'
                            # Используем transform, чтобы сохранить индекс
                            X[result_col] = X.groupby(team_col)['FTR'].transform(
                                lambda x: x.shift(i).map({'H': 1.0, 'D': 0.5, 'A': 0.0} if team_col == 'HomeTeam' 
                                                        else {'A': 1.0, 'D': 0.5, 'H': 0.0})
                            )
                            results.append(result_col)
                            
                            # Для пропущенных голов
                            conceded_col = f'{team_col}_shift{i}_conceded'
                            X[conceded_col] = X.groupby(team_col)['FTAG' if team_col == 'HomeTeam' else 'FTHG'].shift(i)
                            conceded.append(conceded_col)
                        
                        # Заполняем пропуски перед расчетом средних
                        for col in goals + conceded:
                            X[col].fillna(0, inplace=True)  # Если нет данных - считаем 0 голов
                        
                        for col in results:
                            X[col].fillna(0.5, inplace=True)  # Если нет данных - считаем нейтральный результат (0.5)
                        
                        # Расчет средних показателей
                        X[f'{team_col}_last{window}_goals_avg'] = X[goals].mean(axis=1)
                        X[f'{team_col}_last{window}_win_rate'] = X[results].mean(axis=1)
                        X[f'{team_col}_last{window}_conceded_avg'] = X[conceded].mean(axis=1)

            # Разница в форме между командами
            for window in self.window_sizes:
                X[f'form_diff_last{window}'] = (
                    X[f'HomeTeam_last{window}_win_rate'] - 
                    X[f'AwayTeam_last{window}_win_rate']
                )
            
            # Проверка и заполнение возможных NaN в новых признаках
            new_features = list(set(X.columns) - original_columns)
            for feat in new_features:
                if 'win_rate' in feat:
                    X[feat].fillna(0.5, inplace=True)  # Средний результат если нет данных
                elif 'goals' in feat or 'conceded' in feat:
                    X[feat].fillna(0, inplace=True)  # Нет голов если нет данных
            
            logger.info(f"Создано {len(new_features)} новых признаков формы команд")
            return X
        except Exception as e:
            logger.exception(f"Ошибка в TeamFormFeatures.transform: {str(e)}")
            raise
    
# задаем параметры чтобы RandomizedSearchCV сам искал наилучшие параметыр 
def get_param_grid(model):
    if isinstance(model, GradientBoostingRegressor):
        return {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [3, 5, 7],
            'min_samples_leaf': [1, 3]
        }
    elif isinstance(model, XGBRegressor):
        return {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [3, 5],
            'subsample': [0.8, 0.9],
            'gamma': [0, 0.1]
        }
    elif isinstance(model, LGBMRegressor):
        return {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.05],
            'num_leaves': [15, 31],
            'min_child_samples': [10, 20]
        }
    elif isinstance(model, RandomForestRegressor):
        return {
            'n_estimators': [100, 200],
            'max_depth': [5, 10, None],
            'min_samples_leaf': [1, 3, 5]
        }
    elif isinstance(model, HistGradientBoostingRegressor):
        return {
            'max_iter': [100, 200],
            'max_depth': [3, 5],
            'min_samples_leaf': [5, 10],
            'learning_rate': [0.01, 0.05]
        }
    return {}    

def train_football_models(df):
    try:
        # Удаление букмекерских коэффициентов по предсказанию исхода матча
        bookmaker_cols = ['B365H', 'B365D', 'B365A', 'BWH', 'BWD', 'BWA', 'BFH', 'BFD', 'BFA', 
                        'PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'MaxH', 'MaxD', 'MaxA', 
                        'AvgH', 'AvgD', 'AvgA', 'BFEH', 'BFED', 'BFEA', 'AHh', 'B365AHH', 
                        'B365AHA', 'PAHH', 'PAHA', 'MaxAHH', 'MaxAHA', 'AvgAHH', 'AvgAHA', 
                        'BFEAHH', 'BFEAHA', 'B365CH', 'B365CD', 'B365CA', 'BWCH', 'BWCD', 
                        'BWCA', 'BFCH', 'BFCD', 'BFCA', 'PSCH', 'PSCD', 'PSCA', 'WHCH', 
                        'WHCD', 'WHCA', 'MaxCH', 'MaxCD', 'MaxCA', 'AvgCH', 'AvgCD', 'AvgCA', 
                        'BFECH', 'BFECD', 'BFECA', 'AHCh', 'B365CAHH', 'B365CAHA', 'PCAHH', 
                        'PCAHA', 'MaxCAHH', 'MaxCAHA', 'AvgCAHH', 'AvgCAHA', 'BFECAHH', 
                        'BFECAHA', '___________________', '____________']
        bookmaker_cols = [col for col in bookmaker_cols if col in df.columns]
        df = df.drop(bookmaker_cols, axis=1)
        print(df.isnull().sum())

        # Удаление артефактных столбцов
        bad_cols = [col for col in df.columns if col.strip('_') == '']
        if bad_cols:
            df = df.drop(bad_cols, axis=1)
            logger.info(f"Удалено {len(bookmaker_cols)} столбцов с коэффициентами букмекеров")
        
        # Обработка Period и дубликатов
        period_cols = [col for col in df.columns if str(df[col].dtype).startswith('period')]
        if period_cols:
            df[period_cols] = df[period_cols].astype(str)
            logger.info(f"Преобразовано {len(period_cols)} period-столбцов в строки")

        logger.info(f"Исходный размер данных: {df.shape}")    
        
        duplicate_cols = df.columns[df.columns.duplicated()]
        if len(duplicate_cols) > 0:
            df = df.loc[:, ~df.columns.duplicated()]
            logger.info(f"Удалено {len(duplicate_cols)} дубликатов столбцов")

        # Целевые переменные
        y_home = df['FTHG'].astype(float)
        y_away = df['FTAG'].astype(float)

        # Временные признаки
        if 'Date' in df.columns:
            df = df.sort_values('Date')
            df['DaysFromSeasonStart'] = (df['Date'] - df['Date'].min()).dt.days
            df['IsWeekend'] = df['Date'].dt.dayofweek >= 5
            df['Month'] = df['Date'].dt.month
            logger.info("Добавлены временные признаки")

        # Признаки формы команд
        df = TeamFormFeatures().fit_transform(df)

        test_size = 0.2
        split_idx = int(len(df) * (1 - test_size))
        X_train, X_test = df.iloc[:split_idx].copy(), df.iloc[split_idx:].copy()
        y_home_train, y_home_test = y_home.iloc[:split_idx].copy(), y_home.iloc[split_idx:].copy()
        y_away_train, y_away_test = y_away.iloc[:split_idx].copy(), y_away.iloc[split_idx:].copy()

        logger.info(f"Данные разделены: train={len(X_train)}, test={len(X_test)}")

        # Сброс индексов
        X_train.reset_index(drop=True, inplace=True)
        X_test.reset_index(drop=True, inplace=True)
        y_home_train.reset_index(drop=True, inplace=True)
        y_away_train.reset_index(drop=True, inplace=True)

        # Кодирование команд
        if 'HomeTeam' in X_train.columns:
            home_encoder = TargetEncoder(cols=['HomeTeam'], smoothing=10)
            X_train = home_encoder.fit_transform(X_train, y_home_train)  # Теперь индексы совпадают
            X_test = home_encoder.transform(X_test)
            logger.info("Закодированы HomeTeam")
        
        if 'AwayTeam' in X_train.columns:
            away_encoder = TargetEncoder(cols=['AwayTeam'], smoothing=10)
            X_train = away_encoder.fit_transform(X_train, y_away_train)
            X_test = away_encoder.transform(X_test)
            logger.info("Закодированы AwayTeam")

        # Проверка распределений
        if 'HomeTeam' in X_train.columns:
            print("\nПроверка кодирования HomeTeam:")
            print("Train:", X_train['HomeTeam'].describe())
            print("Test:", X_test['HomeTeam'].describe())

        # Приводим название команд к нормальному виду, чтобы не было ошибок
        X_train.columns = X_train.columns.str.replace(r'[^a-zA-Z0-9]', '_', regex=True)
        X_test.columns = X_test.columns.str.replace(r'[^a-zA-Z0-9]', '_', regex=True)
        logger.info("Очищены названия столбцов")

        # Удаление дубликатов после очистки
        duplicate_cols = X_train.columns[X_train.columns.duplicated()]
        if len(duplicate_cols) > 0:
            cols = pd.Series(X_train.columns)
            for dup in cols[cols.duplicated()].unique():
                cols[cols[cols == dup].index] = [f"{dup}_{i}" for i in range(sum(cols == dup))]
            X_train.columns = cols
            X_test.columns = cols

        # Препроцессинг
        numeric_cols = [col for col in X_train.columns if col not in ['Div']]
        categorical_cols = ['Div'] if 'Div' in X_train.columns else []

        preprocessor = ColumnTransformer(
            transformers=[ 
                ('num', make_pipeline(SimpleImputer(strategy='mean'), StandardScaler()), numeric_cols),
                ('cat', make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(handle_unknown='ignore')), categorical_cols)
            ])
        logger.info("Создан препроцессор")

        # Обучение моделей (оригинальные параметры) на основание наилучших показателей
    
        models = {
            'GradientBoosting': GradientBoostingRegressor(n_estimators= 100, min_samples_leaf = 1, max_depth = 3, learning_rate= 0.05, random_state=42),
            'RandomForest': RandomForestRegressor(n_estimators = 200, min_samples_leaf = 1, max_depth = 5, random_state=42),
            'XGBoost': XGBRegressor(subsample = 0.9, n_estimators = 200, max_depth = 3, learning_rate = 0.01, gamma = 0.1, random_state=42),
            'LightGBM': LGBMRegressor(num_leaves = 15, n_estimators = 200, min_child_samples = 10, learning_rate = 0.01, random_state=42),
            'HistGradientBoosting': HistGradientBoostingRegressor(min_samples_leaf = 5, max_iter = 200, max_depth = 3, learning_rate = 0.01, random_state=42)
        }

        logger.info("Инициализированы модели")

        # Обучение и оценка
        def evaluate_model(model, X_train, X_test, y_train, y_test):
            try:
                model.fit(X_train, y_train)
                pred = model.predict(X_test)
                return {
                    'mse': mean_squared_error(y_test, pred),
                    'mae': mean_absolute_error(y_test, pred),
                    'r2': r2_score(y_test, pred),
                    'cv_mse': -cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
                }
            except Exception as e:
                logger.exception(f"Ошибка при оценке модели {model.__class__.__name__}: {str(e)}")
                return None
        
        # Убедимся, что удалили все нечисловые столбцы, кроме целевых
        non_numeric = X_train.select_dtypes(exclude=['number']).columns
        if len(non_numeric) > 0:
            print(f"\nУдаляем нечисловые столбцы: {list(non_numeric)}")
            X_train = X_train.drop(non_numeric, axis=1)
            X_test = X_test.drop(non_numeric, axis=1)

        # Теперь все колонки должны быть числовыми
        numeric_cols = X_train.columns.tolist()
        categorical_cols = []

        preprocessor = make_column_transformer(
            (make_pipeline(SimpleImputer(strategy='mean'), StandardScaler()), numeric_cols),
            (make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(handle_unknown='ignore')), categorical_cols),
            verbose_feature_names_out=False  # Упрощает имена
        )
        '''
        ##############################
        # Инициализация моделей (без параметров)
        models = {
            'GradientBoosting': GradientBoostingRegressor(random_state=42),
            'RandomForest': RandomForestRegressor(random_state=42),
            'XGBoost': XGBRegressor(random_state=42),
            'LightGBM': LGBMRegressor(random_state=42),
            'HistGradientBoosting': HistGradientBoostingRegressor(random_state=42)
        }
        
            # NEW: Модифицированная функция evaluate_model с подбором параметров
        def evaluate_model(model, X_train, X_test, y_train, y_test, n_iter=10):
            param_grid = get_param_grid(model)
            
            if param_grid:  # Если есть параметры для поиска
                print(f"\nПоиск параметров для {model.__class__.__name__}...")
                rs = RandomizedSearchCV(
                    estimator=model,
                    param_distributions=param_grid,
                    n_iter=n_iter,  # Количество итераций для случайного поиска
                    cv=5,
                    scoring='neg_mean_squared_error',
                    n_jobs=-1,
                    verbose=1,
                    random_state=42  # Для воспроизводимости
                )
                rs.fit(X_train, y_train)
                best_model = rs.best_estimator_
                print(f"Лучшие параметры: {rs.best_params_}")
            else:
                best_model = clone(model)
                best_model.fit(X_train, y_train)

            pred = best_model.predict(X_test)
            return {
                'model': best_model,
                'mse': mean_squared_error(y_test, pred),
                'mae': mean_absolute_error(y_test, pred),
                'r2': r2_score(y_test, pred),
                'cv_mse': -cross_val_score(best_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
            }
        
        '''
        #для моделей с параметрам
        def evaluate_model(model, X_train, X_test, y_train, y_test):
            try:
                model = clone(model)
                model.fit(X_train, y_train)
                pred = model.predict(X_test)
                return {
                    'model': model,
                    'mse': mean_squared_error(y_test, pred),
                    'mae': mean_absolute_error(y_test, pred),
                    'r2': r2_score(y_test, pred),
                    'cv_mse': -cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
                }
            except Exception as e:
                logger.exception(f"Ошибка при оценке модели {model.__class__.__name__}: {str(e)}")
                return None

        print("\n=== Обработка данных ===")
        X_train_processed = preprocessor.fit_transform(X_train)
        X_test_processed = preprocessor.transform(X_test)

        # NEW: Сохраняем обученные модели
        trained_models = {}
        home_results = {}
        away_results = {}

        logger.info("Начинаем обучение моделей")
        for name, model in models.items():
            print(f"\n--- Модель: {name} ---")
            home_results[name] = evaluate_model(clone(model), X_train_processed, X_test_processed, y_home_train, y_home_test)
            away_results[name] = evaluate_model(clone(model), X_train_processed, X_test_processed, y_away_train, y_away_test)
            trained_models[name] = {
                'home': home_results[name]['model'],
                'away': away_results[name]['model']
            }
            print(f"Результаты для {name}:")
            print(f"Дома: R2={home_results[name]['r2']:.3f}, MAE={home_results[name]['mae']:.3f}")
            print(f"Гости: R2={away_results[name]['r2']:.3f}, MAE={away_results[name]['mae']:.3f}")

        # NEW: Вывод лучших параметров для каждой обученной модели
        print("\n=== Лучшие параметры для каждой модели ===")
        for name in models.keys():
            if home_results[name]['model'] is not None:
                print(f"{name} (Дома): {home_results[name]['model'].get_params()}")
            if away_results[name]['model'] is not None:
                print(f"{name} (Гости): {away_results[name]['model'].get_params()}")

        # Здесь выводите итоговые результаты
        print("\n=== Итоговые результаты ===")    

        # Создание ансамблей из лучших моделей
        top_home_models = sorted(trained_models.keys(), key=lambda x: home_results[x]['cv_mse'])[:3]
        home_ensemble = StackingRegressor(
            estimators=[(name, trained_models[name]['home']) for name in top_home_models],
            final_estimator=RidgeCV()
        ).fit(X_train_processed, y_home_train)

        top_away_models = sorted(trained_models.keys(), key=lambda x: away_results[x]['cv_mse'])[:3]
        away_ensemble = StackingRegressor(
            estimators=[(name, trained_models[name]['away']) for name in top_away_models],
            final_estimator=RidgeCV()
        ).fit(X_train_processed, y_away_train)

        # Вывод результатов
        print("\n=== Итоговые результаты ===")
        print("{:<25} {:<8} {:<8} {:<8} {:<8} {:<8} {:<8}".format(
            "Модель", "Дома R²", "Гости R²", "Дома MSE", "Гости MSE", "Дома MAE", "Гости MAE"
        ))
        print("-" * 75)
        for name in models.keys():
            print("{:<25} {:<8.3f} {:<8.3f} {:<8.3f} {:<8.3f} {:<8.3f} {:<8.3f}".format(
                name,
                home_results[name]['r2'],
                away_results[name]['r2'],
                home_results[name]['mse'],
                away_results[name]['mse'],
                home_results[name]['mae'],
                away_results[name]['mae']
            ))

        print("\n=== Лучшие модели для ансамбля ===")
        print("Для домашних матчей:", top_home_models)
        print("Для гостевых матчей:", top_away_models)

        return {
            'home_ensemble': home_ensemble,
            'away_ensemble': away_ensemble,
            'preprocessor': preprocessor,
            'home_results': home_results,
            'away_results': away_results,
            'trained_models': trained_models,  # NEW: Все обученные модели
            'feature_names': preprocessor.get_feature_names_out()
        }
    except Exception as e:
        logger.exception(f"Критическая ошибка в train_football_models: {str(e)}")
        raise
# запуск модуля
try:
    logger.info("Запуск анализа футбольных матчей")
    results = train_football_models(df)
    logger.info("Анализ успешно завершен")
except Exception as e:
    logger.exception(f"Фатальная ошибка при выполнении: {str(e)}")
```

```python
def detect_anomalies(df):
    """
    Функция для обнаружения аномалий в данных о футбольных матчах.
    Возвращает DataFrame с оценками аномальности.
    """
    warnings.filterwarnings("ignore", category=RuntimeWarning,
                          module='sklearn.covariance._robust_covariance')
    
    # 1. Подготовка данных
    numeric_features = ['HomeRating', 'AwayRating', 'RatingDiff', 'HomeForm', 'AwayForm', 
                       'H2H_HomeGoals', 'H2H_AwayGoals', 'H2H_WinRate', 'Относительная_форма']
    
    X_anomaly = df[numeric_features].fillna(df[numeric_features].median())
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_anomaly)
    
    # 2. Инициализация моделей обнаружения аномалий
    models = {
        'Isolation Forest': IsolationForest(contamination=0.05, random_state=42),
        'LOF': LocalOutlierFactor(n_neighbors=20, contamination=0.05),
        'Elliptic Envelope': EllipticEnvelope(contamination=0.05, random_state=42, 
                                           support_fraction=0.8),  # Увеличим support_fraction
        'One-Class SVM': OneClassSVM(nu=0.05, kernel="rbf", gamma=0.1),
        'KNN': KNN(contamination=0.05),
        'HBOS': HBOS(contamination=0.05)
    }
    
    # 3. Обучение моделей и вычисление оценок аномальности
    for name, model in models.items():
        try:
            if name == 'LOF':
                # Для LOF используем negative_outlier_factor_
                lof = model.fit(X_scaled)
                df[f'Anomaly_Score_{name}'] = -lof.negative_outlier_factor_
            elif name == 'One-Class SVM':
                model.fit(X_scaled)
                df[f'Anomaly_Score_{name}'] = -model.decision_function(X_scaled)
            else:
                model.fit(X_scaled)
                df[f'Anomaly_Score_{name}'] = -model.decision_function(X_scaled)
        except Exception as e:
            print(f"Ошибка при выполнении модели {name}: {e}")
            df[f'Anomaly_Score_{name}'] = np.nan
    
    # Заполнение пропусков, если какие-то модели не сработали
    anomaly_cols = [col for col in df.columns if 'Anomaly_Score_' in col]
    df[anomaly_cols] = df[anomaly_cols].fillna(df[anomaly_cols].mean())
    
    # 4. Нормализация оценок
    for col in anomaly_cols:
        df[col] = (df[col] - df[col].mean()) / df[col].std()
    
    # 5. Комбинированная оценка (исключаем возможные NaN)
    valid_cols = [col for col in anomaly_cols if not df[col].isnull().any()]
    df['Anomaly_Score_Combined'] = df[valid_cols].mean(axis=1)
    
    # 6. Определение аномалий (верхние 5%)
    threshold = df['Anomaly_Score_Combined'].quantile(0.95)
    df['Is_Anomaly'] = df['Anomaly_Score_Combined'] > threshold
    
    return df, threshold

def analyze_anomalies(df, threshold):
    """
    Анализ и визуализация обнаруженных аномалий (каждый график отдельно)
    """
    anomalies = df[df['Is_Anomaly']]
    normal = df[~df['Is_Anomaly']]
    
    print(f"\nОбнаружено {len(anomalies)} аномальных матчей ({len(anomalies)/len(df)*100:.2f}% от общего числа)")

    # Создаем папку для графиков, если ее нет
    os.makedirs('visualizations/anomaly_plots', exist_ok=True)

    # 1. Распределение комбинированной оценки (отдельный график)
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Anomaly_Score_Combined'], bins=50, kde=True)
    plt.axvline(x=threshold, color='r', linestyle='--', label='Порог аномалий')
    plt.title('Распределение комбинированной оценки аномальности', fontsize=14)
    plt.xlabel('Оценка аномальности', fontsize=12)
    plt.ylabel('Частота', fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.savefig('visualizations/anomaly_plots/score_distribution.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 2. Сравнение методов (отдельный график)
    plt.figure(figsize=(12, 6))
    anomaly_cols = [col for col in df.columns if 'Anomaly_Score_' in col and 'Combined' not in col]
    melted = pd.melt(df, value_vars=anomaly_cols, var_name='Method', value_name='Score')
    sns.boxplot(data=melted, x='Method', y='Score', palette='viridis')
    plt.title('Сравнение оценок аномальности по методам', fontsize=14)
    plt.xlabel('Метод обнаружения аномалий', fontsize=12)
    plt.ylabel('Нормализованная оценка', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('visualizations/anomaly_plots/methods_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 3. Временной анализ аномалий (отдельный график)
    plt.figure(figsize=(14, 6))
    df['Date'] = pd.to_datetime(df['Date'])
    scatter = plt.scatter(df['Date'], df['Anomaly_Score_Combined'], 
                         c=df['Anomaly_Score_Combined'], cmap='coolwarm', alpha=0.7,
                         s=50, edgecolor='k', linewidth=0.5)
    cbar = plt.colorbar(scatter)
    cbar.set_label('Уровень аномальности', fontsize=12)
    plt.axhline(y=threshold, color='r', linestyle='--', linewidth=2)
    plt.title('Аномалии по датам', fontsize=14)
    plt.xlabel('Дата', fontsize=12)
    plt.ylabel('Комбинированная оценка аномальности', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.savefig('visualizations/anomaly_plots/temporal_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 4. Сравнение признаков (отдельный график с исправленными цветами)
    plt.figure(figsize=(16, 8))
    numeric_features = ['HomeRating', 'AwayRating', 'RatingDiff', 'HomeForm', 'AwayForm', 
                       'H2H_HomeGoals', 'H2H_AwayGoals', 'H2H_WinRate', 'Относительная_форма']
    melted = pd.melt(df, id_vars=['Is_Anomaly'], value_vars=numeric_features)
    
    # Преобразуем булевы значения в понятные метки
    melted['Is_Anomaly'] = melted['Is_Anomaly'].map({True: 'Аномальные', False: 'Нормальные'})
    
    # Создаем boxplot с правильными цветами
    ax = sns.boxplot(
        data=melted, 
        x='variable', 
        y='value', 
        hue='Is_Anomaly',
        palette={'Аномальные': '#ff7f0e', 'Нормальные': '#1f77b4'},
        width=0.7,
        showfliers=False
    )
    
    # Настраиваем оформление
    plt.title('Сравнение признаков для аномальных и нормальных матчей', fontsize=14, pad=20)
    plt.xlabel('Признаки', fontsize=12)
    plt.ylabel('Нормализованные значения', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    
    # Улучшаем легенду
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(
        handles, 
        ['Аномальные матчи', 'Нормальные матчи'],
        title='Тип матча',
        bbox_to_anchor=(1.05, 1),
        loc='upper left',
        fontsize=12
    )
    
    # Добавляем сетку
    ax.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.savefig('visualizations/anomaly_plots/features_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 5. Анализ топ-10 аномальных матчей
    top_anomalies = df.nlargest(10, 'Anomaly_Score_Combined')[
        ['Date', 'HomeTeam', 'AwayTeam', 'FTR', 'HomeRating', 'AwayRating', 'RatingDiff', 
         'HomeForm', 'AwayForm', 'Anomaly_Score_Combined']]
    
    print("\nТоп-10 самых аномальных матчей:")
    display(top_anomalies)
    
    # 6. Сохранение результатов
    df.to_csv('processed_data/matches_with_anomalies_improved.csv', index=False)
    print("\nРезультаты сохранены в 'processed_data/matches_with_anomalies_improved.csv'")
    print("Все графики сохранены в папке 'visualizations/anomaly_plots/'")

# Основной блок выполнения
if __name__ == "__main__":

    
    # Обнаружение аномалий
    df_with_anomalies, threshold = detect_anomalies(df)
    
    # Анализ результатов
    analyze_anomalies(df_with_anomalies, threshold)
```
