```python
import psycopg2
import logging

logging.basicConfig(filename='log.log', level=logging.INFO, format="%(asctime)s %(levelname)s %(module)s %(message)s") 
logger = logging.getLogger(__name__)

try:
    conn = psycopg2.connect(dbname="car_price_db", host="192.168.1.196", user="dwhadmin", password="123", port="5432")
except Exception as e:
    logger.exception(f"Ошибка подключения к БД: {e}")

cursor = conn.cursor()
print('Подключение выполнено')
cursor.execute("SELECT count(*) FROM carprice where brand ='Audi'")
carprice_analyst = cursor.fetchall()
print('---count---')
print(carprice_analyst)

cursor.execute("SELECT sum(price) FROM carprice where brand in ('BMW','Audi','Volkswagen')")
carprice_analyst = cursor.fetchall()
print('---sum---')
print(carprice_analyst)

cursor.execute("SELECT ROUND(AVG(\"Year\")) FROM carprice where brand in ('Toyota','Honda')")
carprice_analyst = cursor.fetchall()
print('---AVG---')
for carprice in carprice_analyst:
    print(carprice)    
conn.close()  
```
